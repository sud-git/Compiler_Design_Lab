```python
%%writefile token_classifier.cpp
#include <bits/stdc++.h>
using namespace std;
unordered_set<string> c_keywords = {
    "int","float","return","for","while","struct","char","double",
    "if","else","switch","case","break","continue","void","typedef"
};
unordered_set<string> java_keywords = {
    "class","public","private","protected","static","void","int",
    "float","double","if","else","switch","case","return","new"
};
unordered_set<string> operators_set = {
    "+","-","*","/","==","!=","<=",">=","&&","||","=","<",">"
};
unordered_set<string> separators_set = {
    ";",",","(",")","{","}","[","]"
};
bool isValidIdentifier(const string &token) {
    if(token.empty()) return false;
    if(!(isalpha(token[0]) || token[0]=='_'))
        return false;

    for(char c : token) {
        if(!(isalnum(c) || c=='_'))
            return false;
    }
    return true;
}
bool isInteger(const string &token) {
    if(token.empty()) return false;
    for(char c : token) {
        if(!isdigit(c)) return false;
    }
    return true;
}
bool isFloat(const string &token) {
    bool dot=false;
    if(token.empty()) return false;

    for(char c : token) {
        if(c=='.') {
            if(dot) return false;
            dot=true;
        }
        else if(!isdigit(c))
            return false;
    }
    return dot;
}
bool containsIllegal(const string &token) {
    for(char c : token) {
        if(!(isalnum(c) || c=='_' || c=='.'))
            return true;
    }
    return false;
}
int main() {

    cout<<"Choose Keyword Set:\n";
    cout<<"1. C\n2. Java\n";
    int choice;
    cin>>choice;

    unordered_set<string> keywords =
        (choice==1)? c_keywords : java_keywords;

    cout<<"Enter number of tokens: ";
    int n;
    cin>>n;
    vector<string> tokens(n);
    cout<<"Enter tokens:\n";
    for(int i=0;i<n;i++)
        cin>>tokens[i];
    map<string,int> frequency;
    map<string,int> identifierFreq;
    map<string,int> symbolTable;
    int symbolID=1;
    for(string token : tokens) {

        if(token.empty()) {
            cout<<token<<" -> INVALID_IDENTIFIER (Empty)\n";
            frequency["INVALID"]++;
            continue;
        }
        if(keywords.count(token)) {
            cout<<token<<" -> KEYWORD\n";
            frequency["KEYWORD"]++;
        }
        else if(operators_set.count(token)) {
            cout<<token<<" -> OPERATOR\n";
            frequency["OPERATOR"]++;
        }
        else if(separators_set.count(token)) {
            cout<<token<<" -> SYMBOL\n";
            frequency["SYMBOL"]++;
        }
        else if(isInteger(token)) {
            cout<<token<<" -> NUMERIC_LITERAL (Integer)\n";
            frequency["INTEGER"]++;
        }
        else if(isFloat(token)) {
            cout<<token<<" -> NUMERIC_LITERAL (Float)\n";
            frequency["FLOAT"]++;
        }
        else if(isValidIdentifier(token)) {

            if(keywords.count(token)) {
                cout<<token<<" -> KEYWORD\n";
                frequency["KEYWORD"]++;
            }
            else {
                cout<<token<<" -> VALID_IDENTIFIER\n";
                frequency["IDENTIFIER"]++;
                identifierFreq[token]++;

                if(!symbolTable.count(token))
                    symbolTable[token]=symbolID++;
            }
        }
        else {
            cout<<token<<" -> INVALID_IDENTIFIER (";

            if(isdigit(token[0]))
                cout<<"Starts with digit";
            else if(containsIllegal(token))
                cout<<"Illegal characters";
            else
                cout<<"Unknown rule";

            cout<<")\n";
            frequency["INVALID"]++;
        }
    }
    cout<<"\n===== Frequency Report =====\n";
    for(auto &p : frequency)
        cout<<p.first<<" : "<<p.second<<endl;

    vector<pair<int,string>> idList;
    for(auto &p : identifierFreq)
        idList.push_back({p.second,p.first});

    sort(idList.rbegin(),idList.rend());

    cout<<"\nTop 3 Identifiers:\n";
    for(int i=0;i<min(3,(int)idList.size());i++)
        cout<<idList[i].second<<" ("<<idList[i].first<<")\n";
    cout<<"\n===== Symbol Table =====\n";
    for(auto &p : symbolTable)
        cout<<"ID "<<p.second<<" -> "<<p.first<<endl;

    return 0;
}
```
